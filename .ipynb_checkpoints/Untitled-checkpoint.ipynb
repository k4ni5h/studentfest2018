```python
train_df = train_df.drop(['id', 'test_id', 'age'], axis=1)
test_df = test_df.drop(['test_id', 'age'], axis=1)
combine = [train_df, test_df]
train_df.shape, test_df.shape
```

```python
for dataset in combine:
    dataset['gender'] = dataset['gender'].map( {'F': 1, 'M': 0} ).astype(int)

train_df.head()
```

```python
g = sns.FacetGrid(train_df, col='is_pass')
g.map(plt.hist, 'trainee_id', bins=20)
```

```python
train_df = train_df.drop(['trainee_id'], axis=1)
test_df = test_df.drop(['trainee_id'], axis=1)
combine = [train_df, test_df]
train_df.shape, test_df.shape
```

```python
train_df = train_df.drop(['program_type'], axis=1)
test_df = test_df.drop(['program_type'], axis=1)
combine = [train_df, test_df]
train_df.shape, test_df.shape
```

```python
for dataset in combine:
    dataset['test_type'] = dataset['test_type'].map( {'offline': 1, 'online': 0} ).astype(int)
    dataset['difficulty_level'] = dataset['difficulty_level'].map( {'hard': 3, 'vary hard': 2, 'intermediate': 1, 'easy': 0} ).astype(int)
    dataset['education'] = dataset['education'].map( {'Masters': 4, 'Bachelors': 3, 'High School Diploma': 2, 'Matriculation': 1, 'No Qualification': 0} ).astype(int)
    dataset['is_handicapped'] = dataset['is_handicapped'].map( {'Y': 1, 'N': 0} ).astype(int)

train_df.head()
```

```python
for dataset in combine:
    dataset['trainee_engagement_rating'] = dataset['trainee_engagement_rating'].map( {5.0: 4, 4.0: 3, 3.0: 2, 2.0: 1, 1.0: 0} )
    dataset['city_tier'] = dataset['city_tier'].map( {4: 3, 3: 2, 2: 1, 1: 0} )
```

```python
for dataset in combine:
    dataset['trainee_engagement_rating'] = dataset['trainee_engagement_rating'].fillna(6)
combine[0]['trainee_engagement_rating']=combine[0]['trainee_engagement_rating'].astype(int)
combine[1]['trainee_engagement_rating']=combine[1]['trainee_engagement_rating'].astype(int)
train_df.head()
```

```python
train_df['duration'] = pd.cut(train_df['program_duration'], 8)
train_df[['duration', 'is_pass']].groupby(['duration'], as_index=False).mean().sort_values(by='duration', ascending=True)
```

```python
for dataset in combine:    
    dataset.loc[ dataset['program_duration'] <= 119.375, 'program_duration'] = 0
    dataset.loc[(dataset['program_duration'] > 119.375) & (dataset['program_duration'] <= 121.75), 'program_duration'] = 1
    dataset.loc[(dataset['program_duration'] > 121.75) & (dataset['program_duration'] <= 124.125), 'program_duration'] = 2
    dataset.loc[(dataset['program_duration'] > 128.875) & (dataset['program_duration'] <= 131.25), 'program_duration'] = 3
    dataset.loc[(dataset['program_duration'] > 131.25) & (dataset['program_duration'] <= 133.625), 'program_duration'] = 4
    dataset.loc[ dataset['program_duration'] > 133.625, 'program_duration'] = 5
train_df = train_df.drop(['duration'], axis=1)
combine = [train_df, test_df]
train_df.head()
```

```python
test_df.head()
```

```python
for dataset in combine:
    dataset['program_id'] = dataset['program_id'].map( {'Z_3': 21, 'Z_2': 20, 'Z_1': 19, 'Y_4': 18, 'Y_3': 17, 'Y_2': 16, 'Y_1': 15, 'X_3': 14, 'X_2': 13, 'X_1': 12, 'V_4': 11, 'V_3': 10, 'V_2': 9, 'V_1': 8, 'U_2': 7, 'U_1': 6, 'T_4': 5, 'T_3': 4, 'T_2': 3, 'T_1': 2, 'S_2': 1, 'S_1': 0} ).astype(int)
```

```python
X_train = train_df.drop("is_pass", axis=1)
Y_train = train_df["is_pass"]
X_test  = test_df.drop("id", axis=1).copy()
X_train.shape, Y_train.shape, X_test.shape
```

```python
# Logistic Regression

logreg = LogisticRegression()
logreg.fit(X_train, Y_train)
Y_pred = logreg.predict(X_test)
acc_log = round(logreg.score(X_train, Y_train) * 100, 2)
acc_log
```

```python
coeff_df = pd.DataFrame(train_df.columns.delete(0))
coeff_df.columns = ['Feature']
coeff_df["Correlation"] = pd.Series(logreg.coef_[0])

coeff_df.sort_values(by='Correlation', ascending=False)
```

```python
# Support Vector Machines

svc = SVC()
svc.fit(X_train, Y_train)
Y_pred = svc.predict(X_test)
acc_svc = round(svc.score(X_train, Y_train) * 100, 2)
acc_svc
```

```python
knn = KNeighborsClassifier(n_neighbors = 3)
knn.fit(X_train, Y_train)
Y_pred = knn.predict(X_test)
acc_knn = round(knn.score(X_train, Y_train) * 100, 2)
acc_knn
```

```python
# Gaussian Naive Bayes

gaussian = GaussianNB()
gaussian.fit(X_train, Y_train)
Y_pred = gaussian.predict(X_test)
acc_gaussian = round(gaussian.score(X_train, Y_train) * 100, 2)
acc_gaussian
```

```python
# Perceptron

perceptron = Perceptron()
perceptron.fit(X_train, Y_train)
Y_pred = perceptron.predict(X_test)
acc_perceptron = round(perceptron.score(X_train, Y_train) * 100, 2)
acc_perceptron
```

```python
# Linear SVC

linear_svc = LinearSVC()
linear_svc.fit(X_train, Y_train)
Y_pred = linear_svc.predict(X_test)
acc_linear_svc = round(linear_svc.score(X_train, Y_train) * 100, 2)
acc_linear_svc
```

```python
# Stochastic Gradient Descent

sgd = SGDClassifier()
sgd.fit(X_train, Y_train)
Y_pred = sgd.predict(X_test)
acc_sgd = round(sgd.score(X_train, Y_train) * 100, 2)
acc_sgd
```

```python
# Decision Tree

decision_tree = DecisionTreeClassifier()
decision_tree.fit(X_train, Y_train)
Y_pred = decision_tree.predict(X_test)
acc_decision_tree = round(decision_tree.score(X_train, Y_train) * 100, 2)
acc_decision_tree
```

```python
# Random Forest

random_forest = RandomForestClassifier(n_estimators=100)
random_forest.fit(X_train, Y_train)
Y_pred = random_forest.predict(X_test)
random_forest.score(X_train, Y_train)
acc_random_forest = round(random_forest.score(X_train, Y_train) * 100, 2)
acc_random_forest
```

```python
models = pd.DataFrame({
    'Model': ['Support Vector Machines', 'KNN', 'Logistic Regression', 
              'Random Forest', 'Naive Bayes', 'Perceptron', 
              'Stochastic Gradient Decent', 'Linear SVC', 
              'Decision Tree'],
    'Score': [acc_svc, acc_knn, acc_log, 
              acc_random_forest, acc_gaussian, acc_perceptron, 
              acc_sgd, acc_linear_svc, acc_decision_tree]})
models.sort_values(by='Score', ascending=False)
```
